```python
from numpy import *
import matplotlib.pyplot as plt
import matplotlib.cm as cm

mesh = 1500
eta0 = 0.06
eta1 = 0.112
eta11 = 0.026
angle1 = 32.66*pi/180
angle11 = 42.2*pi/180
beamRadius=1

xx = linspace(-2,2,mesh)
zz = linspace(0,2,mesh)
x,z = meshgrid(xx,zz)

# diameter=1 circle on a 1 x 1 chip
# order1 and order11 are total power for the orders
incident = (abs(x)<=beamRadius**2)
order0 = eta0*(abs(x)<=beamRadius**2)
order1 = eta1*((x-z*tan(angle1))**2<=beamRadius**2)+eta1*((x+z*tan(angle1))**2<=beamRadius**2)+eta1*2*(x**2+(z*tan(angle1))**2<=beamRadius**2)
order11 = 2*eta11*((x-z*tan(angle1))**2+(z*tan(angle1))**2<=beamRadius**2)+2*eta11*((x+z*tan(angle1))**2+(z*tan(angle1))**2<=beamRadius**2)

fig, ax = plt.subplots(1, 1)
mesh = ax.contourf(x, z ,incident-order0-order1-order11,100)
fig.colorbar(mesh)
ax.set_title('ηB vs x and z')
ax.set_xlabel('x(cm)')
ax.set_ylabel('z(cm)')
plt.set_cmap('hsv')
# plt.show()
plt.savefig('balance.png',dpi=700)
```
